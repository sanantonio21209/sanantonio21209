```python
class Solution:
    def findLongestChain(self, pairs) -> int:
        # 一次扫描，对所有数排序，确定矩阵大小
        # 二维矩阵：行数和列数是小数和大数，数值是两者之间的路径长度
        # 进行第二次扫描，每次对三维矩阵进行更新
        # 遍历二维矩阵，找出最大值
        import numpy as np

        numbers = []
        for i in pairs:
            for j in i:
                if j not in numbers:
                    numbers.append(j)
        n = len(numbers)
        numbers = sorted(numbers)
        path = np.zeros((n,n), dtype=np.int)
        for pair in pairs:
            i, j = numbers.index(pair[0]), numbers.index(pair[1])
            if path[i,j] == 0:
                path[i,j] = 1
                for _ in range(j+1, n):
                    for __ in range(_+1, n):
                        if path[_, __]!=0:
                            path[i, __] == max(path[i, __], path[_+1, n]+1)
                for _ in range(0, i):
                    for __ in range(_, i):
                        if path[_, __] != 0:
                            path[_, j] += 1
        return max(map(max, path))

ans = Solution()
print(ans.findLongestChain([[1,2], [3,4], [2,3]]))
print(ans.findLongestChain([[3,4], [2,3]]))
print(ans.findLongestChain([[1,2], [3,4]]))
print(ans.findLongestChain([[1,2], [3,4], [3,5], [4,6], [2,5], [2,4], [1,3], [1,2], [3,6]]))
```

```python
class Solution:
    def uniqueLetterString(self, s: str) -> int:
        
        if len(s)==1: return 1
        
        def order(c: str) -> int:
            return ord(c)-65

        # from enum import Enum
        # class Num(Enum):
        #     ZERO = 0
        #     ONE = 1
        #     TWO = 2

        ans = 0
        import numpy as np
        n = len(s)
        for i in range(n):
            if i==0:
                ans += 1
                continue
            else:
                # char_exist = np.zeros(26, dtype=Num)
                char_exist = np.zeros(26, dtype=np.int8)
                add = 0
                for j in range(i,-1,-1):
                    if char_exist[order(s[j])] == 0:
                        char_exist[order(s[j])] = 1
                        add += 1
                        ans += add
                    elif char_exist[order(s[j])] == 1:
                        char_exist[order(s[j])] = 2
                        add -= 1
                        ans += add
                    else:
                        ans += add
        return ans
    
s = Solution()
s.uniqueLetterString('ABC')



            
```
